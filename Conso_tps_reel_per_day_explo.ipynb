```python
from google.colab import auth
import pandas as pd
auth.authenticate_user()
query = "SELECT * FROM team-elec.Consommation_elec.Conso_tps_reel_per_year"
conso_df = pd.read_gbq(query, project_id = "team-elec")
```

```python
conso_df.head()
```

```python
conso_df_2023=conso_df.iloc[[1]]
conso_df_2023
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Select the columns corresponding to consumption categories
consumption_categories = ['fioul', 'charbon', 'gaz', 'nucleaire', 'eolien', 'solaire', 'hydraulique', 'bioenergies']
consumption_values = conso_df_2023.loc[1, consumption_categories]

# Apply Seaborn styling
sns.set(style="whitegrid")

# Create a pie chart
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(consumption_values, labels=None, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("bright"), explode = [0, 0, 0, 0.1, 0, 0, 0, 0])
plt.legend(
    wedges, consumption_categories, title="Cat√©gories",
    loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)
)
plt.title('Consommation 2023 per Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
